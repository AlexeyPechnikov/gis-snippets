```python
import numpy as np
from IPython.display import HTML
import matplotlib.pyplot as plt
from matplotlib import animation
%matplotlib inline
```

```python
plt.rcParams['figure.figsize'] = (5,3)
plt.rcParams['figure.dpi'] = 100
plt.rcParams['savefig.dpi'] = 100
plt.rcParams["animation.html"] = "jshtml"
```

```python
fig, ax = plt.subplots()
ax.get_yaxis().set_ticks([])
ax.get_xaxis().set_ticks([])

def animate(r):
    mask = !./focal {r}
    mask = [int(v) for v in mask[0][:-1].split(' ')]
    unitmask = np.zeros((2*r+1,2*r+1))
    for (x, y) in zip(mask[::2], mask[1::2]):
        unitmask[x+r,y+r] = 1
    im = ax.imshow(-unitmask, cmap='gray', interpolation='None')
    ax.set_title(f'Radius {r} px', fontsize=18)

ani = animation.FuncAnimation(fig, animate, frames=np.arange(1,10), interval=1000)
plt.close()
writer = animation.PillowWriter(fps=1)
ani.save("focal.gif", writer=writer)
ani
```
